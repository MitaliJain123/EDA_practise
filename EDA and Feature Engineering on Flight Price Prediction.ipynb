```python
## importing Basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
## merge both dataset so whatever we do in one dataset will be done in both
df = pd.concat([train_df, test_df], axis=0)
df.head()
```

```python
df.info()
```

```python
df['Date_of_Journey'].str.split('/')[0] ## getting all the info in split format of date
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]#only date
df['month']=df['Date_of_Journey'].str.split('/').str[1]#only month
df['year']=df['Date_of_Journey'].str.split('/').str[2]#only year
```

```python
df["Date"]=df["Date"].astype(int)
df["month"]=df["month"].astype(int)
df["year"]=df["year"].astype(int)
```
