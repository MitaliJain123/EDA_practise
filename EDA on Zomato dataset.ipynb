### EDA on Zomato Restaurant dataset

```python
#importing useful libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# import the dataset with the help of pandas
df = pd.read_csv("zomato.csv", encoding= 'latin-1')
#to deal with utf-8 error we use encoding which will be of any type here "latin-1"
```

### In data analysis the thing we do are
1. finding missing values
2. explore about the numerical variables
3. explore about categorical variables
4. finding relationship between features

```python
## finding missing values
df.isnull().sum()
## we saw that in Cuisines there are 9 missing values
```

```python
## another way to find the column having missing values
[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
## finding missing values with the help of Heatmap
plt.rcParams['figure.figsize']=(10,10)
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
# since as compare to the whole dataset the null vale is so less that's why we are not able to see that
```

```python
## importing 1 more dataset
df_country = pd .read_excel('Country-Code.xlsx')
df_country.head()
```

```python
## combining 2 datasets using pandas
final_df= pd.merge(df,df_country, on='Country Code', how = 'left')
final_df.head()
```

```python
## another way to check datatypes
final_df.dtypes #like info()
```

```python
final_df.columns
```

```python
## to find records in different countries
final_df.Country.value_counts()
```

```python
Country_names= final_df.Country.value_counts().index
Country_names
```

```python
Country_values = final_df.Country.value_counts().values
Country_values
```

```python
## pie chart
plt.pie(Country_values,labels=Country_names)
```

```python
## since this in this chart all the country names merge with each other
## therefore we are goint to take top 3 countries that uses Zomato
plt.pie(Country_values[:3],labels = Country_names[:3], autopct= '%1.2f%%')
```

### Observation : Zomato maximum records or transaction are from India after the in USA and then in United Kingdom
    

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns= {0:'rating count'})
ratings
```

```python
ratings.head()
```

### observation
1. when rating is between 4.5 to 4.9 ---> Excellent
2. when rating is between 4.0 to 3.4 ---> very good
3. when rating is between 3.5 to 3.9 ---> good
4. when rating is between 3.0 to 3.4 ---> average
5. when rating is between 2.5 to 2.9 ---> average
6. when rating is between 2.0 to 2.4 ---> poor


```python
#visualizing the rating
plt.rcParams['figure.figsize']=(10,10)
sns.barplot(x="Aggregate rating", y="rating count", data= ratings)
#it looks like a gaussian curve
```

```python
plt.rcParams['figure.figsize']=(10,10)
sns.barplot(x="Aggregate rating", y="rating count",hue='Rating color', data= ratings, palette= ['pink','red','orange','yellow','green','blue'])
```

### Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x='Rating color',data= ratings, palette= ['pink','red','orange','yellow','green','blue'] )
```
